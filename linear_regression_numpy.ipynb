```python
import numpy as np  # Importing the NumPy library for numerical operations and array handling
import matplotlib.pyplot as plt  # Importing the Matplotlib library for data visualization and plotting
```

```python
np.random.seed(42) np.random.seed(42)

# Set the random seed for reproducibility of random numbers
X = 2 * np.random.rand(100, 1)  # Generate 100 random numbers in a single column (1D array) scaled to the range [0, 2)
y = 4 + 3 * X + np.random.randn(100, 1)  # Create target values with a linear relationship to X, adding Gaussian noise

plt.scatter(X, y)  # Create a scatter plot of the generated data points (X vs y)
plt.title('Generated Data for Linear Regression')  # Set the title of the plot
plt.xlabel('X')  # Label the x-axis as 'X'
plt.ylabel('y')  # Label the y-axis as 'y'
plt.show()  # Display the plot
```

```python
# Add a column of ones to the input features X to account for the intercept term in the linear regression model
X_b = np.c_[np.ones((100, 1)), X]

# Calculate the best-fit parameters (theta) using the Normal Equation
# This involves computing the inverse of the product of X_b transpose and X_b,
# then multiplying it by X_b transpose and the target variable y
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# Print the intercept term (the first element of theta_best)
print(f"Intercept (00): {theta_best[0][0]}")

# Print the slope term (the second element of theta_best)
print(f"Slope (01): {theta_best[1][0]}")
```

```python
# Create a new array with the input values for prediction
X_new = np.array([[0], [2]])

# Add a bias term (intercept) to the new input data
X_new_b = np.c_[np.ones((2, 1)), X_new]

# Calculate the predicted values using the best theta parameters
y_predict = X_new_b.dot(theta_best)

# Plot the predicted values as a red line
plt.plot(X_new, y_predict, "r", label="Predictions")

# Scatter plot the original data points
plt.scatter(X, y)

# Set the title of the plot
plt.title('Linear Regression Fit')

# Label the x-axis
plt.xlabel('X')

# Label the y-axis
plt.ylabel('y')

# Display the legend for the plot
plt.legend()

# Show the plot
plt.show()
```

```python
def mean_squared_error(y_true, y_pred):
    # Calculate the mean squared error between the true and predicted values
    return np.mean((y_true - y_pred) ** 2)

# Predict the training data using the best theta values and the feature matrix X_b
y_train_predict = X_b.dot(theta_best)

# Calculate the mean squared error for the predictions on the training data
mse = mean_squared_error(y, y_train_predict)

# Print the mean squared error result
print(f"Mean Squared Error (MSE) on Training Data: {mse}")
```
